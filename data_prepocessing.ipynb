## STEP 4: CONVERT CATEGORICAL VALUES TO NUMERICAL VALUES

```python
#Convert name and screen names into lower case 
dataset['name'] = dataset['name'].str.lower()
dataset['screen_name'] = dataset['screen_name'].str.lower()
```

```python
dataset.head(10)
```

```python
#Function to find maximum common subsequence of name and screen_name
def maxSubsequence(screen_name, name): 
    # find the length of the strings 
    m = len(screen_name) 
    n = len(name) 
  
    # declaring the array for storing the dp values 
    L = [[None]*(n + 1) for i in range(m + 1)] 
  
    """Following steps build L[m + 1][n + 1] in bottom up fashion 
    Note: L[i][j] contains length of LCS of X[0..i-1] 
    and Y[0..j-1]"""
    for i in range(m + 1): 
        for j in range(n + 1): 
            if i == 0 or j == 0 : 
                L[i][j] = 0
            elif screen_name[i-1] == name[j-1]: 
                L[i][j] = L[i-1][j-1]+1
            else: 
                L[i][j] = max(L[i-1][j], L[i][j-1]) 
  
    # L[m][n] contains the length of LCS of X[0..n-1] & Y[0..m-1] 
    return L[m][n] 
```

```python
#Check for x in y
x = "ABCD"
y = "ABCABD" 
maxSubsequence(x, y)
```

```python
#Find names with NaN values
faulty_name_index=0
for name in dataset.name:
    print(faulty_name_index,"-",len(name))
    faulty_name_index+=1
```

```python
print(dataset.iloc[faulty_name_index])
```

```python
len(dataset)
```

```python
#Dropping rows where name has a float value
dataset.drop(dataset.index[faulty_name_index],inplace=True)
```

```python
dataset.shape
```

```python
#Finding max subsequence for every (name - screen_name) pair 
max_subsequence = {}
for row in range(len(dataset)):
    name = dataset['name'].iloc[row]
    screen_name = dataset['screen_name'].iloc[row]
    print(row," |- - -| ",name,screen_name)
    
    #find len of max subsequence of screen name in user name and add this len to a list
    max_subsequence[row] = maxSubsequence(screen_name, name)
```

```python
#create a new column having length of max subsequence of screen name in name 
dataset['subseq']= max_subsequence.values()
```

```python
dataset.head()
```

```python
# we can see that names with longer length have more weights. we need to normalize this. [0-1]
# let us divide the len of subsequnce by the  length of the user name

def normaliseNameWeight(row):
    name = row['name']
    subseq = row['subseq']
    return (subseq/len(name))

dataset['name_wt'] = dataset.apply(normaliseNameWeight , axis=1)
dataset.head(10)
```

```python
drop_list=['name','screen_name','subseq']
dataset.drop(drop_list, axis = 1, inplace = True)
```

## STEP 6: FINAL DATASET

```python
#reframe the columns of the dataset

features = ['name_wt','statuses_count', 'followers_count', 'friends_count','favourites_count','listed_count','label']

final_dataset = pd.DataFrame(dataset, columns = features)
final_dataset.head()
```

```python
#shuffling rows
final_dataset=final_dataset.sample(frac=1)
final_dataset.head()
```
